```python
def task_plot(task_type, num_task: int):
	fig, axes = plt.subplots(1, num_task)
	fig.set_size_inches(25, 5)
	for task_id in range(num_task):
		if task_type == 'small':
			task = load_small_task(task_id)
		elif task_type == 'medium':
			task = load_medium_task(task_id)
		else:
			task = load_large_task(task_id)
		plant_x = []
		plant_y = []
		dc_x = []
		dc_y = []
		retailer_x = []
		retailer_y = []
		customer_x = []
		customer_y = []
		for i in range(task.num_plants):
			plant_x.append(task.lst_plants[i].coord_x)
			plant_y.append(task.lst_plants[i].coord_y)
		for i in range(task.num_dcs):
			dc_x.append(task.lst_dcs[i].coord_x)
			dc_y.append(task.lst_dcs[i].coord_y)
		for i in range(task.num_retailers):
			retailer_x.append(task.lst_retailers[i].coord_x)
			retailer_y.append(task.lst_retailers[i].coord_y)
		for i in range(task.num_customers):
			customer_x.append(task.lst_customers[i].coord_x)
			customer_y.append(task.lst_customers[i].coord_y)
		axes[task_id].scatter(plant_x, plant_y, label='Plant')
		axes[task_id].scatter(dc_x, dc_y, label='DC')
		axes[task_id].scatter(retailer_x, retailer_y, label='Retailer')
		axes[task_id].scatter(customer_x, customer_y, label='Customer')
		axes[task_id].legend(loc='upper right')
	fig.tight_layout()
	

	
```

```python
task_plot("small", 5)
```

```python
task_plot("medium", 5)
```

```python
task_plot("large", 5)
```
